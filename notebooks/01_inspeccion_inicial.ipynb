```python
import pandas as pd
import json

ruta_archivo = 'dataset_PI.json'

try:
    with open(ruta_archivo, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    df = pd.DataFrame(data)
    
    print("=== INSPECCIÓN INICIAL DEL DATASET ===")
    print(f"Dimensiones del dataset: {df.shape[0]} filas y {df.shape[1]} columnas\n")
    
    print("--- Tipos de Datos y Valores Nulos ---")
    
    resumen = pd.DataFrame({
        'Tipo de Dato': df.dtypes,
        'Valores Nulos': df.isnull().sum(),
        '% de Nulos': (df.isnull().sum() / len(df) * 100).round(2)
    })
    display(resumen)
    
    print("\n--- Vista Previa de los Datos ---")
    display(df.head())
    
    print("\n--- Estadísticas Descriptivas (Numéricas) ---")
    display(df.describe().round(2))
    
    print("\n--- Estadísticas Descriptivas (Categóricas) ---")
    display(df.describe(include=['object', 'category']))

except Exception as e:
    print(f"Error al cargar el archivo: {e}")
```
