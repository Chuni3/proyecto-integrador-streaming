```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


# Configuración visual para que los gráficos se vean profesionales
sns.set_theme(style="whitegrid")

# 1. Cargar el dataset procesado y limpio
ruta_procesado = 'dataset_limpio.csv'
df = pd.read_csv(ruta_procesado)

print(f"Dataset cargado exitosamente. Filas: {df.shape[0]}, Columnas: {df.shape[1]}")

# ==========================================
# ANÁLISIS UNIVARIADO: Distribución de Planes
# ==========================================
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x='subscription_plan', palette='viridis', order=df['subscription_plan'].value_counts().index)

plt.title('Distribución de Usuarios por Plan de Suscripción', fontsize=14)
plt.xlabel('Plan de Suscripción', fontsize=12)
plt.ylabel('Cantidad de Usuarios', fontsize=12)

# Agregar las cantidades arriba de cada barra
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

grafico analisis univariado: en el grafico poddemos visualizar cual es el plan de subscripcion mas contratado. 

```python
# ==========================================
# GRÁFICO 2 (UNIVARIADO 2): Histograma de Edades
# ==========================================
plt.figure(figsize=(9, 5))
sns.histplot(data=df, x='age', bins=20, kde=True, color='skyblue')

plt.title('Distribución de la Edad de los Usuarios', fontsize=14, fontweight='bold')
plt.xlabel('Edad (Años)', fontsize=12)
plt.ylabel('Cantidad de Usuarios (Frecuencia)', fontsize=12)
plt.show()
```

grafico de analisis univariado,: aqui muestro la la poblacion de edad de los clientes, donde predomina las personas de entre 30 y 40 años

```python
# ==========================================
# GRÁFICO 3 (BIVARIADO 1 ALTERNATIVO): Promedio de Minutos por Plan
# ==========================================
plt.figure(figsize=(8, 5))

# sns.barplot por defecto calcula el promedio (mean) de la variable numérica
sns.barplot(data=df, x='subscription_plan', y='monthly_watch_time_mins', palette='Set2', errorbar=None)

plt.title('Promedio de Minutos Visualizados al Mes por Plan', fontsize=14, fontweight='bold')
plt.xlabel('Plan de Suscripción', fontsize=12)
plt.ylabel('Promedio de Minutos / Mes', fontsize=12)

plt.show()
```

grafico bivariado: en este grafico vemos el promedio de minutos de visualizacion segun el plan contratado por los clientes

```python
# ==========================================
# GRÁFICO 4 (BIVARIADO 2): Minutos Promedio por Género Favorito
# ==========================================
plt.figure(figsize=(10, 5))

# Calculamos el promedio de minutos para ordenar las barras de mayor a menor
orden_generos = df.groupby('favorite_genre')['monthly_watch_time_mins'].mean().sort_values(ascending=False).index

sns.barplot(data=df, x='favorite_genre', y='monthly_watch_time_mins', palette='viridis', order=orden_generos, errorbar=None)

plt.title('Promedio de Minutos de Visualización Mensual por Género Favorito', fontsize=14, fontweight='bold')
plt.xlabel('Género Favorito', fontsize=12)
plt.ylabel('Minutos Promedio / Mes', fontsize=12)
plt.xticks(rotation=15) # Rotamos los nombres para que se lean bien
plt.show()
```

grafico bivariado: aqui vemos el promedio de visualizacion en minutos de cada genero favorito por los clientes

```python
# ==========================================
# GRÁFICO 5 (MULTIVARIADO): Edad vs. Minutos por Plan de Suscripción
# ==========================================
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='monthly_watch_time_mins', hue='subscription_plan', alpha=0.7, palette='deep')

plt.title('Relación entre Edad y Minutos de Visualización según el Plan', fontsize=14, fontweight='bold')
plt.xlabel('Edad (Años)', fontsize=12)
plt.ylabel('Minutos Visualizados al Mes', fontsize=12)
plt.legend(title='Plan de Suscripción')
plt.show()
```

grafico multivariado: aqui podemos visualizar la poblacion de clientes segun la edad, los minutos de visualizacion mensual y el plan de cada uno
